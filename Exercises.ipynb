```
# get summary marital status rates by occupation
result = spark.sql(
  """
  SELECT
    occupation,
    SUM(1) as n,
    ROUND(AVG(if(LTRIM(marital_status) LIKE 'Married-%',1,0)),2) as married_rate,
    ROUND(AVG(if(lower(marital_status) LIKE '%widow%',1,0)),2) as widow_rate,
    ROUND(AVG(if(LTRIM(marital_status) LIKE 'Divorced',1,0)),2) as divorce_rate,
    ROUND(AVG(if(LTRIM(marital_status) LIKE 'Separated',1,0)),2) as separated_rate,
    ROUND(AVG(if(LTRIM(marital_status) LIKE 'Never-married',1,0)),2) as bachelor_rate
  FROM
    adult
  GROUP BY 1
  ORDER BY n DESC
  """)

display(result)
```

```
#register df as a table
sqlContext.registerDataFrameAsTable(result, "result")
spark.sql("SELECT * FROM result").show(5)
```

```
#similar calculation we did above
import pyspark.sql.functions as F

#wrangle data
df_result = df_adult.select(
  df_adult['occupation'],
  F.when(F.ltrim(F.col('marital_status')) == 'Divorced',1).otherwise(0).alias('is_divorced')
)

#grouping
df_result = df_result.groupBy('occupation').agg(F.round(F.mean('is_divorced'),2).alias('divorced_rate'))
#ordering
df_result = df_result.orderBy(F.desc('divorced_rate'))

#show results
df_result.show(10)
```

```
from pyspark.ml import Pipeline
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler

categoricalColumns = ["workclass", "education", "marital_status", "occupation", "relationship", "race", "sex",
                     "native_country"]
stages = [] #stages in our pipeline

for i in categoricalColumns:
  #Category indexing
  stringIndexer = StringIndexer(inputCol=i, outputCol=i + "Index")
  # Use OneHotEncoder to convert categorical variables into binary SparseVectors
  # encoder = OneHotEncoder(inputCol=categoricalCol + "Index", outputCol=categoricalCol + "classVec")
  encoder = OneHotEncoder(inputCols=[stringIndexer.getOutputCol()], outputCols=[i + "classVec"])
  stages += [stringIndexer, encoder]
  
#Convert label into label indices using StringIndexer
label_StringIdx = StringIndexer(inputCol="income", outputCol="label")
stages += [label_StringIdx]

#Transform all features into a vector using VecotorAssembler
numericCols = ["age", "fnlwgt", "education_num", "capital_gain", "capital_loss", "hours_per_week"]
assemblerInputs = [c + "classVec" for c in categoricalColumns] + numericCols
assembler = VectorAssembler(inputCols=assemblerInputs, outputCol="features")
stages += [assembler]

#Create a Pipeline
pipeline  = Pipeline(stages=stages)
#Run the feature transformations
#  - fit() computes feature statistics as needed.
#  - transform() actually transforms the features.
pipelineModel = pipeline.fit(df_adult)
dataset = pipelineModel.transform(df_adult)
#Keep relevant columns
selectedcols = ["label", "features"] + cols
dataset = dataset.select(selectedcols)
#display(dataset)
dataset.show(30)
```

```
#Ramdomly split data into training and test sets. Set seed for reproducibility
(trainingData, testData) = dataset.randomSplit([0.7, 0.3], seed=100)
print(trainingData.count())
print(testData.count())
```

```
#get the rate of the postiive outcome from the training data to use as a threshhold in the model
training_data_positive_rate = trainingData.select(F.avg(trainingData['label'])).collect()[0][0]

print("Positive rate in the training data is {}".format(training_data_positive_rate))
```

```
from pyspark.ml.classification import LogisticRegression

# Create initial LogisticRegression model
lr = LogisticRegression(labelCol="label", featuresCol="features", maxIter=10)

# set threshold for the probability above which to predict a 1
lr.setThreshold(training_data_positive_rate)
# lr.setThreshold(0.5) # could use this if knew you had balanced data

# Train model with Training Data
lrModel = lr.fit(trainingData)

# get training summary used for eval metrics and other params
lrTrainingSummary = lrModel.summary

# Find the best model threshold if you would like to use that instead of the empirical positve rate
fMeasure = lrTrainingSummary.fMeasureByThreshold
maxFMeasure = fMeasure.groupBy().max('F-Measure').select('max(F-Measure)').head()
lrBestThreshold = fMeasure.where(fMeasure['F-Measure'] == maxFMeasure['max(F-Measure)']) \
    .select('threshold').head()['threshold']
  
print("Best threshold based on model performance on training data is {}".format(lrBestThreshold))
```
